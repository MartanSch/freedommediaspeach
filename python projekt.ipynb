```python
import pandas as pd
import numpy as np
```

```python
freedomhouse = pd.read_csv('FHdataset2.csv', sep = ';', on_bad_lines= 'skip')
freedomhouse
```

```python
# d1 free and independent media
# d2 express and practice religion
# d3 academic freedom and free from politics education
# d4 free to express political views
```

```python
data = freedomhouse[['Country/Territory', 'Edition', 'D1' ,'D4', 'Total']]
data.head()
```

```python
data = data.rename(columns = {
    'D1' : 'media',
    'D4' : 'speech'
})
```

```python
data.head(6)
```

```python
data_media_total = data[['media','Total']]
data_media_total.head()
```

```python
data_media_total.corr()
```

```python
data_speech_total = data[['speech','Total']]
data_speech_total.head()
```

```python
data_speech_total.corr()
```

```python
data_speech_media = data[['speech','media']]
data_speech_media.head()
```

```python
data_speech_media.corr()
```

```python
data_media_over_speech = data.loc[data['media'] > data['speech']]
data_media_over_speech
```

```python
a = round(data_media_over_speech.shape[0]/data.shape[0],4)
print(f'only in {a*100}% of cases the media score is higher than speech score')
```
